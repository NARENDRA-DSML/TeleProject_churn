Model 1 - Logistics Regression

```python
from sklearn.linear_model import LogisticRegression
# create instance of LR
logM1 = LogisticRegression(max_iter=100)

#train the model
logM1.fit(x_train, y_train)
y_pred = logM1.predict(x_test)
# define the accuracy of the model
from sklearn import metrics
acc=metrics.accuracy_score(y_test, y_pred)
cnf=metrics.confusion_matrix(y_test, y_pred)
cnf
```

Model 2 - Statsmodel.api - Training Dataset

```python
import statsmodels.api as sm
x_train_const = sm.add_constant(x_train)
logM2 = sm.GLM(y_train, x_train_const, family=sm.families.Binomial())
result = logM2.fit()
y_pred = result.predict(x_train_const)
```

```python
y_pred = y_pred.apply(lambda x:1 if x>0.5 else 0)
y_pred.head()
```

```python
acc1=metrics.accuracy_score(y_train, y_pred)
cnf1=metrics.confusion_matrix(y_train, y_pred)
cnf1
```

```python
acc1
```

```python
# test Dataset
x_test_const = sm.add_constant(x_test)
y_pred_test = result.predict(x_test_const)
y_pred_test = y_pred_test.apply(lambda x:1 if x>0.5 else 0)
acc2= metrics.accuracy_score(y_test, y_pred_test)
```

#Recursive Feature Elimination

```python
# it develops a model by checking existence of feature 1
# first we train a machine learning model
# remove the unnecessary attributes
# retrain the model
# repeat the process of removing until we get a satisfactory result
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
from sklearn.feature_selection import RFE
rfe = RFE(logreg)
rfe.fit(x_train, y_train)
```

```python
#true indicates feature is important
# false indicates not important
rfe.support_
```

```python
# no of important features
rfe.n_features_
```

```python
rfe.n_features_in_
```

```python
rfe.feature_names_in_
```

```python
rfe.ranking_
```

```python
list(zip(x_train.columns, rfe.ranking_, rfe.support_))
```

```python
# list or display the important features
col_imp = x_train.columns[rfe.support_]
col_imp
```

```python
# list or display the unimportant features by using ~v operator
col_unimp = x_train.columns[~rfe.support_]
col_unimp
```

```python
# create the model using important features
x_train_imp = sm.add_constant(x_train[col_imp])
logM3 = sm.GLM(y_train, x_train_imp, family=sm.families.Binomial())
resultimp = logM3.fit()
y_pred_imp = resultimp.predict(x_train_imp)
```

```python
y_pred_imp = y_pred_imp.apply(lambda x:1 if x>0.5 else 0)
acc2=metrics.accuracy_score(y_train, y_pred_imp)
cnf2=metrics.confusion_matrix(y_train, y_pred_imp)
cnf2
```

```python
acc2
```

```python
x_test_imp = sm.add_constant(x_test[col_imp])
y_pred_test_imp = resultimp.predict(x_test_imp)
y_pred_test_imp = y_pred_test_imp.apply(lambda x:1 if x>0.5 else 0)
acc3= metrics.accuracy_score(y_test, y_pred_test_imp)
acc3
```
